```python
import seaborn as sns
import matplotlib.pyplot as plt

ct = pd.crosstab(
    crime_df['Heads of Crime'],
    crime_df['Major Heads']
)

plt.figure(figsize=(12,6))
sns.heatmap(ct, annot=True, fmt='d')
plt.title("Crime Relationship Heatmap")
plt.show()
```

```python
from scipy.stats import chi2_contingency

ct = pd.crosstab(crime_df['Heads of Crime'], crime_df['Major Heads'])

chi2, p, dof, expected = chi2_contingency(ct)

print("p-value:", p)
```

```python
crime_df['MoM_change'] = crime_df['During the current month'] - crime_df['During the previous month']
crime_df['MoM_change']
```

```python
mom_increase = crime_df.sort_values(by='MoM_change', ascending=False).head(10)
print(mom_increase[['Major Heads', 'Minor Heads', 'MoM_change']])
```

```python
mom_decrease = crime_df.sort_values(by='MoM_change').head(10)
mom_decrease
```

```python
crime_df['YoY_change'] = crime_df['During the current month'] - crime_df['During the corresponding month of previous year']
```

```python
yoy_increase = crime_df.sort_values(by='YoY_change', ascending=False).head(10)
yoy_increase
```

```python
yoy_decrease = crime_df.sort_values(by='YoY_change').head(10)
yoy_decrease
```

```python
crime_df['MoM_%'] = (crime_df['MoM_change'] / crime_df['During the previous month'].replace(0,1)) * 100
crime_df['YoY_%'] = (crime_df['YoY_change'] / crime_df['During the corresponding month of previous year'].replace(0,1)) * 100
```

```python
major_trend = crime_df.groupby('Major Heads')[
    ['During the current month','During the previous month','During the corresponding month of previous year']
].sum()

major_trend['MoM_change'] = major_trend['During the current month'] - major_trend['During the previous month']
major_trend['YoY_change'] = major_trend['During the current month'] - major_trend['During the corresponding month of previous year']

major_trend = major_trend.sort_values(by='During the current month', ascending=False)

print(major_trend.head(10))
```

```python
import matplotlib.pyplot as plt

top = major_trend.head(5)

top[['During the current month','During the previous month']].plot(kind='bar', figsize=(10,5))
plt.title("Current vs Previous Month")
plt.xticks(rotation=45)
plt.show()
```

```python
top[['During the current month','During the corresponding month of previous year']].plot(kind='bar', figsize=(10,5))
plt.title("Current vs Last Year")
plt.xticks(rotation=45)
plt.show()
```
